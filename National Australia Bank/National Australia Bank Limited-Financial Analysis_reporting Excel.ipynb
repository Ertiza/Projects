```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
%matplotlib notebook 

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/113.0.0.0 Safari/537.36',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8',
    'Accept-Language': 'en-US,en;q=0.5',
    'Connection': 'keep-alive',
    'Upgrade-Insecure-Requests': '1',
    'Cache-Control': 'max-age=0'
}

ticker = 'NABZY'

urls = {}
urls['income_annually'] = f"https://stockanalysis.com/stocks/{ticker}/financials/"
# urls['income_quarterly'] = f"https://stockanalysis.com/stocks/{ticker}/financials/?p=quarterly"
urls['balancesheet_annually'] = f"https://stockanalysis.com/stocks/{ticker}/financials/balance-sheet/"
# urls['balancesheet_quarterly'] = f"https://stockanalysis.com/stocks/{ticker}/financials/balance-sheet/?p=quarterly"
urls['cashflow_annually'] = f"https://stockanalysis.com/stocks/{ticker}/financials/cash-flow-statement/"

# xlwriter = pd.ExcelWriter(f'financial_statements({ticker}).xlsx', engine='xlsxwriter')

dataframes = {}

# Iterate over the URLs and retrieve data into DataFrames
for key, url in urls.items():
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')
    table = soup.find('table')  # Assuming the data is present in a table

    # Convert the table to a pandas DataFrame
    dataframe = pd.read_html(str(table))[0]  # Assumes only one table is present
    dataframes[key] = dataframe

# Now you have dataframes containing the data from each URL
# You can access them using the keys defined in the `urls` dictionary

# Example: Accessing the income_annually DataFrame
income_statement = dataframes['income_annually']
freeCash_Flow = dataframes['cashflow_annually']
balance_sheet = dataframes['balancesheet_annually']
```

## Creating Excel sheet for data storing and Reporting - with charts

```python
import pandas as pd

def create_sheet(writer, df, sheet_name):
    df.to_excel(writer, sheet_name=sheet_name, index_label='Year')
    worksheet = writer.sheets[sheet_name]
    workbook = writer.book

    chart = workbook.add_chart({'type': 'line'})

    chart.add_series({
        'name': sheet_name,
        'categories': [sheet_name, 1, 0, 11, 0],
        'values': [sheet_name, 1, 1, 11, 1],
       
    })
    column_chart = workbook.add_chart({'type': 'column'})

    column_chart.add_series({
        'name': sheet_name,
        'categories': [sheet_name, 1, 0, 11, 0],
        'values': [sheet_name, 1, 2, 11, 2],
        'y2_axis' : True,
    })
    chart.combine(column_chart)
    
    
    chart.set_title({'name': sheet_name})
    chart.set_x_axis({'name': 'Date'})
    chart.set_y_axis({'name': 'value'})
    column_chart.set_y2_axis({'name': '%-chg'})

    worksheet.insert_chart("E1", chart)

    
```

```python
# Create a Pandas ExcelWriter object
writer = pd.ExcelWriter("National_Australia_Bank_Limited_financials.xlsx", engine='xlsxwriter')
workbook = writer.book

# Write DataFrames to sheets
create_sheet(writer, rev, 'Revenue')
create_sheet(writer, eps, 'EPS')
create_sheet(writer, fcf, 'FCF')
create_sheet(writer, WkgCptl, 'Working_Capital')

# Close the workbook
workbook.close()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your dataset is stored in a DataFrame named 'data'

# Convert necessary columns to numeric type
data['Revenue'] = pd.to_numeric(data['Revenue'], errors='coerce')
data['EPS'] = pd.to_numeric(data['EPS'], errors='coerce')
data['FCF'] = pd.to_numeric(data['FCF'], errors='coerce')
data['WkgCptl'] = pd.to_numeric(data['WkgCptl'], errors='coerce')

data = data.sort_index(ascending=False)

# Set the figure size
fig, ax1 = plt.subplots(figsize=(8,4))

# Plot Revenue
ax1.plot(data.index, data['Revenue'], color='orange', label='Revenue')
ax1.set_xlabel('Index')
ax1.set_ylabel('Revenue', color='orange')
ax1.tick_params(axis='y', labelcolor='orange')
ax1.set_title('Financial Performance', fontsize=16)

# Create a second y-axis for EPS
ax2 = ax1.twinx()
ax2.plot(data.index, data['EPS'], color='blue', label='EPS')
ax2.set_ylabel('EPS', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

# Create a third y-axis for FCF
ax3 = ax1.twinx()
ax3.spines['right'].set_position(('outward', 60))
ax3.plot(data.index, data['FCF'], color='green', label='FCF')
ax3.set_ylabel('FCF', color='green')
ax3.tick_params(axis='y', labelcolor='green')

# Create a fourth y-axis for Working Capital
ax4 = ax1.twinx()
ax4.spines['right'].set_position(('outward', 120))
ax4.plot(data.index, data['WkgCptl'], color='purple', label='Working Capital')
ax4.set_ylabel('Working Capital', color='purple')
ax4.tick_params(axis='y', labelcolor='purple')

# Combine the legends from all axes
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
lines3, labels3 = ax3.get_legend_handles_labels()
lines4, labels4 = ax4.get_legend_handles_labels()
ax1.legend(lines + lines2 + lines3 + lines4, labels + labels2 + labels3 + labels4, loc='best')

# Reverse the x-axis order
ax1.set_xlim(ax1.get_xlim()[::-1])

# Adjust the layout and spacing
plt.tight_layout()

# Show the chart
plt.show()
```
